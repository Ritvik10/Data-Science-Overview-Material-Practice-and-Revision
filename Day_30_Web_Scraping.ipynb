# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from bs4 import BeautifulSoup
import requests
```

# Web Scraping

```python
url = "https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
req = requests.get(url)
```

```python
req.status_code
```

```python
req
```

```python
soup = BeautifulSoup(req.content)
```

```python
soup.prettify()
```

```python
arr = [i for i in soup.findAll('td')]
```

```python
arr[0]
```

```python
arr[2]
```

```python
re.sub('^<td>.*">|<td>|</td>|<.*>|\n','',str(arr[3]))
```

```python
Film = []
Year = []
Award = []
Nomination = []
```

```python
count = 0

for i in soup.findAll('td'):
  new_i = re.sub('^<td>.*">|<td>|</td>|<.*>|\n','',str(i))
  if count == 0:
    Film.append(new_i)
    count += 1
  elif count == 1:
    Year.append(new_i)
    count += 1
  elif count == 2:
    Award.append(new_i)
    count += 1
  else:
    Nomination.append(new_i)
    count = 0
```
