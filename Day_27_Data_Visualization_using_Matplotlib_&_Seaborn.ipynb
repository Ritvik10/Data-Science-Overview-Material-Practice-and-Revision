# Line Chart

```python
x = np.array([101, 102, 103, 104, 105])
y = np.array([0, 10, 9, 2, 20])
```

```python
x,y
```

```python
plt.plot(x,y)
```

* Making a line chart without line

```python
plt.plot(x,y,'o')
```

```python
plt.plot(x,y,'d')
```

```python
# References of markers
"""
'o' -> Circle
'd' -> Diamond
'*' -> Star
'h' -> Hexagon
"""
```

```python
# References of Lines
"""
- -> Solid Line
: -> Dotted line
-- -> Dashed Line
-. -> Dashed-Dotted line
"""
```

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,ls=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle='-.')
```

* Format String -> fmt(marker, line, color)

```python
plt.plot(y,x)
```

```python
plt.plot(y,x,'o:r')
```

```python
# References of color
"""
r -> Red
g -> Green
b -> Blue
y -> Yellow
w -> White
k -> Black
"""
```

```python
plt.plot(x,y,'d--y')
```

* Labels

```python
x = [1, 2, 3, 4]
y = [2, 5, 7, 11]
```

```python
plt.plot(x,y)

plt.xlabel('Height')
plt.ylabel('Weight')
```

* Title

```python
plt.plot(x,y)

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height Vs Weight')
```

```python
modification = {'color': 'red', 'size':20, 'family':'serif'}

plt.plot(x,y)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.title('Height Vs Weight', fontdict = modification, loc='left')
```

* Subplots -> Multiple plots in a single canvas

```python
# 1st plot
x = [1,2,3,4,5,6]
y = [10, 20, 30, 40, 50, 60]
# subplot(row, column, plot_number)
plt.title('Age')
plt.subplot(2,3,1)
plt.plot(x,y)


# 2nd plot
x = [8, 3, 1, 10]
y = [2, 6, 7, 11]
plt.subplot(2,3,2)
plt.plot(x,y)



# 3rd plot
x = [1,2,3,4,5,6]
y = [10, 20, 30, 40, 50, 60]
# subplot(row, column, plot_number)
plt.subplot(2,3,3)
plt.plot(x,y)


# 4th plot
x = [8, 3, 1, 10]
y = [2, 6, 7, 11]
plt.subplot(2,3,4)
plt.plot(x,y)




# 5th plot
x = [1,2,3,4,5,6]
y = [10, 20, 30, 40, 50, 60]
# subplot(row, column, plot_number)
plt.subplot(2,3,5)
plt.plot(x,y)


# 6th plot
x = [8, 3, 1, 10]
y = [2, 6, 7, 11]
plt.subplot(2,3,6)
plt.plot(x,y)

plt.suptitle('Weather Condition')
```

# Piechart

```python
x = [40, 20, 30, 10]
```

```python
plt.pie(x, labels=["Apples", 'Bananas', "Pear", "Grapes"],
        colors = ["red", 'yellow', 'green', 'black'],
        shadow = True,
        explode = [0, 0.1, 0, 0])

plt.legend(title='Fruits')
```
