* Creating a DataFrame from a Dict

```python
data = {"Height":[70.8, 90.5, 56, 23, 51],
        "Weight":[34, 78, 55.7, 120, 72],
        "Age":[10, 20, 30, 40, 50]}


df = pd.DataFrame(data)
df
```

```python
data = {"Height":[70.8, 90.5, 56, 23, 51],
        "Weight":[34, 78, 55.7, 120, 72],
        "Age":[10, 20, 30, 40, 50, 60, 70]}


df = pd.DataFrame(data)
df
```

```python
data = {"Height":[70.8, 90.5, 56, 23, 51],
        "Weight":[34, 78, 55.7, 120, 72],
        "Age":[10, 20, 30, 40, 50]}


df = pd.DataFrame(data, columns=["H", "W", "A"])
df
```

```python
data = {"Height":[70.8, 90.5, 56, 23, 51],
        "Weight":[34, 78, 55.7, 120, 72],
        "Age":[10, 20, 30, 40, 50]}


df = pd.DataFrame(data, index=["A", "B", "C", "D", "E","F"])
df
```

* Creating a DataFrame from a List of Dicts

```python
data = [{"Height": 56,
         "Weight": 89,
         "Age": 98,
         "Name": "Steve"},
        {"Height": 78,
         "Weight": 89,
         "Age": 98,
         "Name": "Tony"},
        {"Height": 56,
         "Weight": 89,
         "Age": 98,
         "Name": "Bruce",
         "Email": "xyz@xyz.xyz"},
        {"Country": "India"}]

df = pd.DataFrame(data)
df
```

* Creating a DataFrame from a Dict of Series

```python
data = {"A": pd.Series([10, 20, 30, 40, 50]),
        "B": pd.Series([34, 45, 56, 67, 78]),
        "C": pd.Series([67, 89, 12, 34, 56])}

df = pd.DataFrame(data)
df
```
