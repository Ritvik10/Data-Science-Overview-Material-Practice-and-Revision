# Data Cleaning

```python
data = np.random.randn(5,3)
df = pd.DataFrame(data,
                  index = ['A', 'C', 'E', 'F', 'G'],
                  columns = ['Col1', 'Col2', 'Col3'])
```

```python
df
```

```python
# reindex()
df = df.reindex(["A", "B", "C", "D", "E", "F", "G", "H"])
```
